```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix
from torch.utils.data import DataLoader, TensorDataset
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
import time
import seaborn as sns
from imblearn.over_sampling import SMOTE
import os

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Define save directory in /content/
save_dir = '/content/models'
os.makedirs(save_dir, exist_ok=True)

# ==============================================
# Data Loading and Preprocessing
# ==============================================

# Load the dataset
data = pd.read_csv('balanced.csv')

# Drop columns with all zeros and other unnecessary columns
columns_to_drop = [
    'pslist.nprocs64bit',
    'handles.nport',
    'psxview.not_in_eprocess_pool',
    'psxview.not_in_eprocess_pool_false_avg',
    'svcscan.interactive_process_services',
    'callbacks.nanonymous',
    'Family'  # Not needed for binary classification
]

data = data.drop(columns_to_drop, axis=1)

# Filter to only include Benign and Ransomware samples
data = data[data['Category'].isin(['Benign', 'Ransomware'])]

# For binary classification, we'll use Category as target (Benign vs Ransomware)
# We can drop the Class column since it combines all malware types
data = data.drop(['Class'], axis=1)

# Convert categorical columns
categorical_columns = ['Category']
label_encoder = LabelEncoder()
for col in categorical_columns:
    data[col] = label_encoder.fit_transform(data[col])

# Normalization
scaler = StandardScaler()
columns_to_scale = [col for col in data.columns if col != 'Category']
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])

# ==============================================
# Data Preparation
# ==============================================

# Separate features and labels
X = data.drop('Category', axis=1).values
y = data['Category'].values

# Apply SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# Convert to PyTorch tensors
X = torch.tensor(X_smote, dtype=torch.float32)
y = torch.tensor(y_smote, dtype=torch.long)

# ==============================================
# LSTM Model Definition
# ==============================================

class LSTMModel(nn.Module):
    def __init__(self, input_size, lstm_hidden_size, num_lstm_layers, num_classes):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, lstm_hidden_size, num_lstm_layers,
                           batch_first=True, dropout=0.2)
        self.dropout = nn.Dropout(0.2)
        self.fc = nn.Linear(lstm_hidden_size, num_classes)

    def forward(self, x):
        # Add sequence dimension (batch_size, sequence_len=1, features)
        x = x.unsqueeze(1)
        lstm_out, _ = self.lstm(x)
        lstm_out = self.dropout(lstm_out[:, -1, :])  # Take the last output
        output = self.fc(lstm_out)
        return output

# ==============================================
# Training and Evaluation Function
# ==============================================

def train_and_evaluate(X, y, num_epochs=20):
    # Define the number of folds for cross-validation
    num_splits = 5
    kf = StratifiedKFold(n_splits=num_splits, shuffle=True, random_state=42)

    # Initialize metrics storage
    fold_accuracies = []
    train_times = []
    test_times = []
    all_true_labels = []
    all_predicted_labels = []

    # Model hyperparameters
    input_size = X.shape[1]  # Number of features
    lstm_hidden_size = 64    # Reduced for binary classification
    num_lstm_layers = 1      # Simpler model for binary task
    num_classes = len(torch.unique(y))

    # Loop over folds
    for fold, (train_index, test_index) in enumerate(kf.split(X, y)):
        print(f"\n=== Fold {fold + 1}/{num_splits} ===")

        # Split data
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Create DataLoaders
        train_dataset = TensorDataset(X_train, y_train)
        test_dataset = TensorDataset(X_test, y_test)
        batch_size = 64  # Increased batch size
        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
        test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

        # Initialize model, loss, and optimizer
        model = LSTMModel(input_size, lstm_hidden_size, num_lstm_layers, num_classes)
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=3)

        # Training loop
        train_losses = []
        test_losses = []
        train_accuracies = []
        test_accuracies = []

        for epoch in range(num_epochs):
            start_time = time.time()
            model.train()
            correct_train = 0
            total_train = 0
            running_train_loss = 0.0

            for inputs, labels in train_loader:
                optimizer.zero_grad()
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()

                running_train_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                total_train += labels.size(0)
                correct_train += (predicted == labels).sum().item()

            # Training metrics
            train_accuracy = correct_train / total_train
            train_loss = running_train_loss / len(train_loader)
            train_losses.append(train_loss)
            train_accuracies.append(train_accuracy)

            # Evaluation
            model.eval()
            correct_test = 0
            total_test = 0
            running_test_loss = 0.0

            with torch.no_grad():
                for inputs, labels in test_loader:
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                    running_test_loss += loss.item()

                    _, predicted = torch.max(outputs.data, 1)
                    total_test += labels.size(0)
                    correct_test += (predicted == labels).sum().item()

                    # Store predictions for confusion matrix
                    if epoch == num_epochs - 1:  # Only last epoch
                        all_true_labels.extend(labels.cpu().numpy())
                        all_predicted_labels.extend(predicted.cpu().numpy())

            # Test metrics
            test_accuracy = correct_test / total_test
            test_loss = running_test_loss / len(test_loader)
            test_losses.append(test_loss)
            test_accuracies.append(test_accuracy)

            # Adjust learning rate
            scheduler.step(test_loss)

            # Timing
            epoch_time = time.time() - start_time
            if epoch == 0:
                train_times.append(epoch_time)
            if epoch == num_epochs - 1:
                test_times.append(epoch_time)

            # Print progress
            if (epoch + 1) % 5 == 0 or epoch == 0 or epoch == num_epochs - 1:
                print(f"Epoch {epoch + 1}/{num_epochs} - "
                      f"Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, "
                      f"Train Acc: {train_accuracy:.4f}, Test Acc: {test_accuracy:.4f}")

        # Save the model after training for this fold
        model_path = os.path.join(save_dir, f'lstm_model_fold_{fold + 1}_epochs_{num_epochs}.pth')
        torch.save(model.state_dict(), model_path)
        print(f"Model saved to {model_path}")

        # Store fold accuracy
        fold_accuracies.append(test_accuracies[-1])

    # ==============================================
    # Results Analysis
    # ==============================================

    print("\n=== Final Results ===")
    print(f"Mean Accuracy: {np.mean(fold_accuracies) * 100:.2f}%")
    print(f"Mean Training Time: {np.mean(train_times):.2f} sec/epoch")
    print(f"Mean Testing Time: {np.mean(test_times):.2f} sec/epoch")

    # Plot training history for last fold
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(range(num_epochs), train_accuracies, label="Train")
    plt.plot(range(num_epochs), test_accuracies, label="Test")
    plt.title(f"Accuracy (Epochs={num_epochs})")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(range(num_epochs), train_losses, label="Train")
    plt.plot(range(num_epochs), test_losses, label="Test")
    plt.title(f"Loss (Epochs={num_epochs})")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Confusion matrix
    conf_matrix = confusion_matrix(all_true_labels, all_predicted_labels)
    plt.figure(figsize=(6, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Benign', 'Ransomware'],
                yticklabels=['Benign', 'Ransomware'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()

# ==============================================
# Run Experiments with Different Epochs
# ==============================================

epoch_settings = [20, 100, 150]

for epochs in epoch_settings:
    print(f"\n\n=== Training with {epochs} Epochs ===")
    train_and_evaluate(X, y, num_epochs=epochs)
```

```python
import os
import pandas as pd
import torch
import torch.nn as nn
import numpy as np
from sklearn.preprocessing import StandardScaler
import pickle
import time
from collections import deque
import matplotlib.pyplot as plt

# Load the saved model and preprocessing objects
model_path = '/content/models/lstm_model_fold_1_epochs_20.pth'  # Using the first fold model
scaler_path = '/content/drive/MyDrive/ransomware_models/scaler.pkl'
label_encoder_path = '/content/drive/MyDrive/ransomware_models/label_encoder.pkl'

# Load model architecture parameters (assuming you saved them)
class LSTMModel(nn.Module):
    def __init__(self, input_size, lstm_hidden_size, num_lstm_layers, num_classes):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, lstm_hidden_size, num_lstm_layers,
                            batch_first=True, dropout=0.2)
        self.dropout = nn.Dropout(0.2)
        self.fc = nn.Linear(lstm_hidden_size, num_classes)

    def forward(self, x):
        x = x.unsqueeze(1)
        lstm_out, _ = self.lstm(x)
        lstm_out = self.dropout(lstm_out[:, -1, :])
        return self.fc(lstm_out)

# ✅ Set input_size = 49 to match your trained model
input_size = 49
model = LSTMModel(input_size=input_size, lstm_hidden_size=64, num_lstm_layers=1, num_classes=2)
model.load_state_dict(torch.load(model_path))
model.eval()

# Load scaler and label encoder
with open(scaler_path, 'rb') as f:
    scaler = pickle.load(f)

with open(label_encoder_path, 'rb') as f:
    label_encoder = pickle.load(f)

# ==============================================
# Behavioral Analysis System
# ==============================================

class BehavioralAnalyzer:
    def __init__(self, model, scaler, window_size=10, anomaly_threshold=0.8):
        self.model = model
        self.scaler = scaler
        self.window_size = window_size
        self.anomaly_threshold = anomaly_threshold
        self.prediction_history = deque(maxlen=window_size)
        self.confidence_history = deque(maxlen=window_size)
        self.timestamps = deque(maxlen=window_size)

    def preprocess(self, new_data):
        required_cols = [col for col in new_data.columns if col != 'Category']
        if len(required_cols) != input_size:
            raise ValueError(f"Expected {input_size} features, but got {len(required_cols)}")
        new_data = new_data[required_cols]
        scaled_data = self.scaler.transform(new_data)
        return torch.tensor(scaled_data, dtype=torch.float32)

    def predict_with_confidence(self, input_data):
        with torch.no_grad():
            outputs = self.model(input_data)
            probabilities = torch.softmax(outputs, dim=1)
            confidence, predicted = torch.max(probabilities, 1)
        print("Input shape to model:", input_data.shape)  # Debug line
        return predicted.item(), confidence.item()

    def analyze_behavior(self, new_sample):
        input_tensor = self.preprocess(new_sample)
        prediction, confidence = self.predict_with_confidence(input_tensor)
        label = label_encoder.inverse_transform([prediction])[0]

        timestamp = time.strftime('%Y-%m-%d %H:%M:%S')
        self.prediction_history.append(prediction)
        self.confidence_history.append(confidence)
        self.timestamps.append(timestamp)

        analysis = {
            'timestamp': timestamp,
            'prediction': label,
            'confidence': confidence,
            'is_anomaly': False,
            'behavior_shift': False,
            'warning': None
        }

        if confidence < self.anomaly_threshold:
            analysis['is_anomaly'] = True
            analysis['warning'] = f"Low confidence prediction ({confidence:.2f})"

        if len(self.prediction_history) == self.window_size:
            unique, counts = np.unique(self.prediction_history, return_counts=True)
            mode_count = max(counts)

            if mode_count / self.window_size < 0.7:
                analysis['behavior_shift'] = True
                analysis['warning'] = "Behavioral shift detected - inconsistent predictions"

            ransomware_count = sum(1 for p in self.prediction_history if p == 1)
            if ransomware_count / self.window_size > 0.6:
                analysis['warning'] = "Potential ransomware activity detected"

        return analysis

    def plot_confidence_trend(self):
        if len(self.confidence_history) < 2:
            print("Not enough data for trend analysis")
            return

        plt.figure(figsize=(10, 5))
        plt.plot(self.timestamps, self.confidence_history, marker='o')
        plt.axhline(y=self.anomaly_threshold, color='r', linestyle='--', label='Anomaly Threshold')
        plt.title('Prediction Confidence Over Time')
        plt.xlabel('Timestamp')
        plt.ylabel('Confidence Score')
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

# ==============================================
# Simulation of Real-time Monitoring
# ==============================================

def simulate_dynamic_monitoring(test_samples_path, analyzer, num_samples=50):
    test_data = pd.read_csv(test_samples_path)
    test_samples = test_data.sample(num_samples) if num_samples < len(test_data) else test_data

    print("\n=== Starting Dynamic Monitoring Simulation ===")
    print(f"Processing {len(test_samples)} samples...\n")

    results = []
    for idx, (_, row) in enumerate(test_samples.iterrows()):
        sample = pd.DataFrame([row])
        analysis = analyzer.analyze_behavior(sample)

        if analysis['warning']:
            print(f"\nALERT at {analysis['timestamp']}:")
            print(f"- Prediction: {analysis['prediction']}")
            print(f"- Confidence: {analysis['confidence']:.2f}")
            print(f"- Warning: {analysis['warning']}")

        results.append(analysis)

        if idx > 0 and idx % 10 == 0:
            analyzer.plot_confidence_trend()

    return results

# ==============================================
# Execution
# ==============================================

if __name__ == '__main__':
    analyzer = BehavioralAnalyzer(model, scaler, window_size=15, anomaly_threshold=0.85)
    new_data_path = '/content/drive/MyDrive/new_samples.csv'  # ✅ Update this path if needed
    monitoring_results = simulate_dynamic_monitoring(new_data_path, analyzer, num_samples=50)

    results_df = pd.DataFrame(monitoring_results)
    results_path = '/content/drive/television_detected'
    results_df.to_csv(results_path, index=False)
    print(f"\nMonitoring results saved to {results_path}")
```

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
from torch.utils.data import DataLoader, TensorDataset
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
import time
import seaborn as sns
from imblearn.over_sampling import SMOTE
import os

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Define save directory in /content/
save_dir = '/content/models'
os.makedirs(save_dir, exist_ok=True)

# ==============================================
# Data Loading and Preprocessing
# ==============================================

# Load the dataset
data = pd.read_csv('balanced.csv')

# Drop columns with all zeros and other unnecessary columns
columns_to_drop = [
    'pslist.nprocs64bit',
    'handles.nport',
    'psxview.not_in_eprocess_pool',
    'psxview.not_in_eprocess_pool_false_avg',
    'svcscan.interactive_process_services',
    'callbacks.nanonymous',
    'Family'  # Not needed for binary classification
]

data = data.drop(columns_to_drop, axis=1)

# Filter to only include Benign and Ransomware samples
data = data[data['Category'].isin(['Benign', 'Ransomware'])]

# For binary classification, we'll use Category as target (Benign vs Ransomware)
data = data.drop(['Class'], axis=1)

# Convert categorical columns
categorical_columns = ['Category']
label_encoder = LabelEncoder()
for col in categorical_columns:
    data[col] = label_encoder.fit_transform(data[col])

# Normalization
scaler = StandardScaler()
columns_to_scale = [col for col in data.columns if col != 'Category']
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])

# ==============================================
# Data Preparation
# ==============================================

# Separate features and labels
X_np = data.drop('Category', axis=1).values
y_np = data['Category'].values

# Apply SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X_np, y_np)

# Convert to PyTorch tensors
X = torch.tensor(X_smote, dtype=torch.float32)
y = torch.tensor(y_smote, dtype=torch.long)

# ==============================================
# LSTM Model Definition
# ==============================================

class LSTMModel(nn.Module):
    def __init__(self, input_size, lstm_hidden_size, num_lstm_layers, num_classes):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, lstm_hidden_size, num_lstm_layers,
                           batch_first=True, dropout=0.2)
        self.dropout = nn.Dropout(0.2)
        self.fc = nn.Linear(lstm_hidden_size, num_classes)

    def forward(self, x):
        x = x.unsqueeze(1)
        lstm_out, _ = self.lstm(x)
        lstm_out = self.dropout(lstm_out[:, -1, :])
        output = self.fc(lstm_out)
        return output

# ==============================================
# Training and Evaluation Function
# ==============================================

def train_and_evaluate(X, y, num_epochs=20):
    num_splits = 5
    kf = StratifiedKFold(n_splits=num_splits, shuffle=True, random_state=42)

    fold_accuracies = []
    train_times = []
    test_times = []
    all_true_labels = []
    all_predicted_labels = []

    input_size = X.shape[1]
    lstm_hidden_size = 64
    num_lstm_layers = 1
    num_classes = len(torch.unique(y))

    for fold, (train_index, test_index) in enumerate(kf.split(X, y)):
        print(f"\n=== Fold {fold + 1}/{num_splits} ===")

        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        train_dataset = TensorDataset(X_train, y_train)
        test_dataset = TensorDataset(X_test, y_test)
        batch_size = 64
        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
        test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

        model = LSTMModel(input_size, lstm_hidden_size, num_lstm_layers, num_classes)
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=3)

        train_losses = []
        test_losses = []
        train_accuracies = []
        test_accuracies = []

        for epoch in range(num_epochs):
            start_time = time.time()
            model.train()
            correct_train = 0
            total_train = 0
            running_train_loss = 0.0

            for inputs, labels in train_loader:
                optimizer.zero_grad()
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()

                running_train_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                total_train += labels.size(0)
                correct_train += (predicted == labels).sum().item()

            train_accuracy = correct_train / total_train
            train_loss = running_train_loss / len(train_loader)
            train_losses.append(train_loss)
            train_accuracies.append(train_accuracy)

            model.eval()
            correct_test = 0
            total_test = 0
            running_test_loss = 0.0

            with torch.no_grad():
                for inputs, labels in test_loader:
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                    running_test_loss += loss.item()

                    _, predicted = torch.max(outputs.data, 1)
                    total_test += labels.size(0)
                    correct_test += (predicted == labels).sum().item()

                    if epoch == num_epochs - 1:
                        all_true_labels.extend(labels.cpu().numpy())
                        all_predicted_labels.extend(predicted.cpu().numpy())

            test_accuracy = correct_test / total_test
            test_loss = running_test_loss / len(test_loader)
            test_losses.append(test_loss)
            test_accuracies.append(test_accuracy)

            scheduler.step(test_loss)

            epoch_time = time.time() - start_time
            if epoch == 0:
                train_times.append(epoch_time)
            if epoch == num_epochs - 1:
                test_times.append(epoch_time)

            if (epoch + 1) % 5 == 0 or epoch == 0 or epoch == num_epochs - 1:
                print(f"Epoch {epoch + 1}/{num_epochs} - "
                      f"Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, "
                      f"Train Acc: {train_accuracy:.4f}, Test Acc: {test_accuracy:.4f}")

        model_path = os.path.join(save_dir, f'lstm_model_fold_{fold + 1}_epochs_{num_epochs}.pth')
        torch.save(model.state_dict(), model_path)
        print(f"Model saved to {model_path}")

        fold_accuracies.append(test_accuracies[-1])

    print("\n=== Final Results ===")
    print(f"Mean Accuracy: {np.mean(fold_accuracies) * 100:.2f}%")
    print(f"Mean Training Time: {np.mean(train_times):.2f} sec/epoch")
    print(f"Mean Testing Time: {np.mean(test_times):.2f} sec/epoch")

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(range(num_epochs), train_accuracies, label="Train")
    plt.plot(range(num_epochs), test_accuracies, label="Test")
    plt.title(f"Accuracy (Epochs={num_epochs})")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(range(num_epochs), train_losses, label="Train")
    plt.plot(range(num_epochs), test_losses, label="Test")
    plt.title(f"Loss (Epochs={num_epochs})")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.tight_layout()
    plt.show()

    conf_matrix = confusion_matrix(all_true_labels, all_predicted_labels)
    precision = precision_score(all_true_labels, all_predicted_labels)
    recall = recall_score(all_true_labels, all_predicted_labels)
    f1 = f1_score(all_true_labels, all_predicted_labels)

    print(f"\nPrecision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {f1:.4f}")

    plt.figure(figsize=(6, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Benign', 'Ransomware'],
                yticklabels=['Benign', 'Ransomware'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()

# ==============================================
# Run Experiments with Different Epochs
# ==============================================

epoch_settings = [20, 100, 150]

for epochs in epoch_settings:
    print(f"\n\n=== Training with {epochs} Epochs ===")
    train_and_evaluate(X, y, num_epochs=epochs)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (roc_curve, auc, precision_recall_curve,
                            confusion_matrix, classification_report,
                            accuracy_score, precision_score, recall_score,
                            f1_score, roc_auc_score)
from sklearn.model_selection import StratifiedKFold
import torch
import os

# Set style for plots (fallback to default if seaborn not available)
try:
    plt.style.use('seaborn')
except:
    plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# ==============================================
# 1. Load Model and Data for Visualization
# ==============================================

def load_predictions(model_path, X_test, y_test, input_size):
    """Load model and get predictions with error handling"""
    try:
        model = LSTMModel(input_size=input_size,
                         lstm_hidden_size=64,
                         num_lstm_layers=1,
                         num_classes=2)
        model.load_state_dict(torch.load(model_path))
        model.eval()

        with torch.no_grad():
            outputs = model(X_test)
            _, predicted = torch.max(outputs.data, 1)
            probabilities = torch.softmax(outputs, dim=1)[:, 1]  # Probability of ransomware class

        return predicted.numpy(), probabilities.numpy(), y_test.numpy()
    except Exception as e:
        print(f"Error loading model or making predictions: {e}")
        return None, None, None

# Load the trained model (using 150 epoch model from fold 1)
model_path = '/content/models/lstm_model_fold_1_epochs_150.pth'

# Get test data from the last fold
try:
    kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    train_index, test_index = list(kf.split(X_np, y_np))[-1]  # Using last fold's test set
    X_test = torch.tensor(X_np[test_index], dtype=torch.float32)
    y_test = torch.tensor(y_np[test_index], dtype=torch.long)

    # Get predictions
    y_pred, y_prob, y_true = load_predictions(model_path, X_test, y_test, input_size=X_np.shape[1])

    if y_pred is None:
        raise ValueError("Failed to get predictions")
except Exception as e:
    print(f"Error preparing test data: {e}")
    raise

# ==============================================
# 2. Diagnostic Checks
# ==============================================

print("\n=== Data Diagnostics ===")
print(f"Number of test samples: {len(y_true)}")
print(f"Class distribution in test set: {np.bincount(y_true)}")
print(f"Unique predicted classes: {np.unique(y_pred)}")
print(f"Probability range: [{np.min(y_prob):.4f}, {np.max(y_prob):.4f}]")

# ==============================================
# 3. Visualization Functions with Error Handling
# ==============================================

def plot_roc_curve(y_true, y_prob):
    """Plot ROC curve with error handling"""
    try:
        if len(np.unique(y_true)) < 2:
            raise ValueError("Only one class present in y_true")

        fpr, tpr, _ = roc_curve(y_true, y_prob)
        roc_auc = auc(fpr, tpr)

        plt.figure()
        plt.plot(fpr, tpr, color='darkorange', lw=2,
                label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic')
        plt.legend(loc="lower right")
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Could not plot ROC curve: {e}")

def plot_pr_curve(y_true, y_prob):
    """Plot Precision-Recall curve with error handling"""
    try:
        if len(np.unique(y_true)) < 2:
            raise ValueError("Only one class present in y_true")

        precision, recall, _ = precision_recall_curve(y_true, y_prob)

        # Ensure recall is sorted in increasing order
        if not np.all(np.diff(recall) >= 0):
            sort_idx = np.argsort(recall)
            recall = recall[sort_idx]
            precision = precision[sort_idx]

        pr_auc = auc(recall, precision)

        plt.figure()
        plt.plot(recall, precision, color='blue', lw=2,
                label=f'PR curve (AUC = {pr_auc:.2f})')
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision-Recall Curve')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.legend(loc="lower left")
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Could not plot Precision-Recall curve: {e}")

def plot_confusion_matrix(y_true, y_pred):
    """Plot confusion matrix with error handling"""
    try:
        cm = confusion_matrix(y_true, y_pred)

        plt.figure(figsize=(6, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=['Benign', 'Ransomware'],
                   yticklabels=['Benign', 'Ransomware'])
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Could not plot confusion matrix: {e}")

def print_classification_report(y_true, y_pred):
    """Print classification report with error handling"""
    try:
        report = classification_report(y_true, y_pred,
                                     target_names=['Benign', 'Ransomware'])
        print("Classification Report:")
        print(report)
    except Exception as e:
        print(f"Could not generate classification report: {e}")

def plot_feature_importance(model, feature_names):
    """Plot feature importance with error handling"""
    try:
        # Get the weights from the first LSTM layer
        lstm_weights = model.lstm.weight_ih_l0.detach().numpy()
        # Sum absolute weights across all hidden units
        importance = np.sum(np.abs(lstm_weights), axis=0)

        # Create a DataFrame for visualization
        importance_df = pd.DataFrame({
            'Feature': feature_names,
            'Importance': importance
        }).sort_values('Importance', ascending=False)

        # Plot top 15 features
        plt.figure(figsize=(10, 8))
        sns.barplot(x='Importance', y='Feature', data=importance_df.head(15))
        plt.title('Top 15 Important Features (Based on LSTM Layer Weights)')
        plt.xlabel('Importance Score')
        plt.ylabel('Feature')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Could not plot feature importance: {e}")

def plot_threshold_analysis(y_true, y_prob):
    """Plot threshold analysis with error handling"""
    try:
        if len(np.unique(y_true)) < 2:
            raise ValueError("Only one class present in y_true")

        thresholds = np.linspace(0, 1, 50)
        f1_scores = []
        precisions = []
        recalls = []

        for thresh in thresholds:
            y_pred_thresh = (y_prob >= thresh).astype(int)
            f1_scores.append(f1_score(y_true, y_pred_thresh, zero_division=0))
            precisions.append(precision_score(y_true, y_pred_thresh, zero_division=0))
            recalls.append(recall_score(y_true, y_pred_thresh, zero_division=0))

        plt.figure(figsize=(10, 6))
        plt.plot(thresholds, f1_scores, label='F1 Score', lw=2)
        plt.plot(thresholds, precisions, label='Precision', lw=2)
        plt.plot(thresholds, recalls, label='Recall', lw=2)
        plt.xlabel('Threshold')
        plt.ylabel('Score')
        plt.title('Threshold Analysis')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Could not plot threshold analysis: {e}")

def create_metrics_summary(y_true, y_pred, y_prob):
    """Create metrics summary table with error handling"""
    try:
        def safe_metric(metric_func, *args, **kwargs):
            try:
                return metric_func(*args, **kwargs)
            except Exception as e:
                print(f"Warning: Could not compute {metric_func.__name__}: {str(e)}")
                return float('nan')

        metrics = {
            'Accuracy': safe_metric(accuracy_score, y_true, y_pred),
            'Precision': safe_metric(precision_score, y_true, y_pred, zero_division=0),
            'Recall': safe_metric(recall_score, y_true, y_pred, zero_division=0),
            'F1-Score': safe_metric(f1_score, y_true, y_pred, zero_division=0),
            'ROC-AUC': safe_metric(roc_auc_score, y_true, y_prob),
            'PR-AUC': safe_metric(lambda yt, yp: auc(*precision_recall_curve(yt, yp)[:2]), y_true, y_prob)
        }

        df = pd.DataFrame.from_dict(metrics, orient='index', columns=['Value'])

        plt.figure(figsize=(8, 4))
        plt.axis('off')
        plt.table(cellText=df.round(4).values,
                 rowLabels=df.index,
                 colLabels=df.columns,
                 cellLoc='center',
                 loc='center',
                 colColours=['#f5f5f5']*len(df.columns))
        plt.title('Final Model Performance Metrics', y=1.1)
        plt.tight_layout()
        plt.show()

        return df
    except Exception as e:
        print(f"Could not create metrics summary: {e}")
        return None

# ==============================================
# 4. Generate All Visualizations
# ==============================================

print("\n=== ROC Curve ===")
plot_roc_curve(y_true, y_prob)

print("\n=== Precision-Recall Curve ===")
plot_pr_curve(y_true, y_prob)

print("\n=== Confusion Matrix ===")
plot_confusion_matrix(y_true, y_pred)

print("\n=== Classification Report ===")
print_classification_report(y_true, y_pred)

print("\n=== Feature Importance ===")
feature_names = data.drop('Category', axis=1).columns.tolist()
model = LSTMModel(input_size=X_np.shape[1], lstm_hidden_size=64, num_lstm_layers=1, num_classes=2)
model.load_state_dict(torch.load(model_path))
plot_feature_importance(model, feature_names)

print("\n=== Threshold Analysis ===")
plot_threshold_analysis(y_true, y_prob)

print("\n=== Final Metrics Summary ===")
final_metrics = create_metrics_summary(y_true, y_pred, y_prob)
if final_metrics is not None:
    print("\nDetailed Metrics:")
    print(final_metrics.round(4))
```
